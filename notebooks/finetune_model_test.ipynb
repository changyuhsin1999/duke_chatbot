```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig

ft_model_id = "cindy990915/duke_chatbot0413merged"
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

ft_model = AutoModelForCausalLM.from_pretrained(
    ft_model_id,  # Mistral, same as before
    quantization_config=bnb_config,  # Same quantization config as before
    device_map="auto",
    trust_remote_code=True,
    use_auth_token=True
)

tokenizer = AutoTokenizer.from_pretrained(ft_model_id, add_bos_token=True, trust_remote_code=True)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

def respond(query):
    eval_prompt = """User Query:\n\n {} ###\n\n""".format(query)
    model_input = tokenizer(eval_prompt, return_tensors="pt").to("cuda")
    output = ft_model.generate(input_ids=model_input["input_ids"].to(device),
                           attention_mask=model_input["attention_mask"],
                           max_new_tokens=125, repetition_penalty=1.15)
    result = tokenizer.decode(output[0], skip_special_tokens=True).replace(eval_prompt, "")
    return result
```

```python
respond("What do well-run software companies have?")
```

```python
respond("What is the role of software architecture in enterprise technology system?")
```
